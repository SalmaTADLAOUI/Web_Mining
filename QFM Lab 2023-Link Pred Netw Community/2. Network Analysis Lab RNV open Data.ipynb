## Definitions

Edges and nodes are the basic building blocks of a network/graph:
- A **node**, also known as a vertex, represents an entity within the network, 
- An **edge**, also known as a link or connection, represents the relationship or interaction between nodes.

Mathematically, a network can be represented as a graph, which is a collection of nodes and edges. A **graph** can be defined as G = (V, E), where V represents the set of nodes or vertices, and E represents the set of edges or connections between them.

Each edge in the graph can be described by a pair of nodes (u, v), where u and v are nodes in V. This represents a directed edge from u to v, which indicates that there is a relationship or interaction from u to v. If the edge is undirected, it can be represented as an unordered pair {u, v}, which indicates that the relationship or interaction is bidirectional.

# Weighted undirected Graph

This is a simple social network graph example, where the **nodes** are the actors and the **edges(undirected)** indicate the relationships between each other. We assign a **weight** to each node representing how many messages two actors have exchanged.

![weighted_undirected.png](https://drive.google.com/uc?id=1fGrBiJDc3jB3ekOjOxcuOjPsuW3mlRYS)

````python
Actor_nw_undirected_weight = nx.Graph()
# Add edges from list and assign weight
Actor_nw_undirected_weight.add_edges_from([('Emma','Shane'), ('Leah','Shane'), ('Bob','John'), ('Emma','Mike')], weight=1)
Actor_nw_undirected_weight.add_edges_from([('Allen','Lisa'), ('Jill','Mike'), ('Liz','Shane')], weight=2)
Actor_nw_undirected_weight.add_edges_from([('Bob','Mike'), ('John','Shane')], weight=3)
Actor_nw_undirected_weight.add_edges_from([('Allen','Liz'), ('Jill','John')], weight=4)
Actor_nw_undirected_weight.add_edges_from([('Bob','Jill')], weight=5)
Actor_nw_undirected_weight.add_edges_from([('Jill','Leah')], weight=6)
Actor_nw_undirected_weight.add_edges_from([('John','Leah')], weight=7)
Actor_nw_undirected_weight.add_edges_from([('Bob','Emma')], weight=8)
Actor_nw_undirected_weight.add_edges_from([('Jill','Shane')], weight=10)
Actor_nw_undirected_weight.add_edges_from([('Emma','Jill')], weight=15)
Actor_nw_undirected_weight.add_edges_from([('Emma','Liz')], weight=16)
````

Networkx offers multiple options for drawing graphs.
In this tutorial we use a [circular layout](https://networkx.org/documentation/stable/reference/generated/networkx.drawing.layout.circular_layout.html) to plot the nodes.

````python
import matplotlib.pyplot as plt

# Extract weight and edge information
weights = [(d['weight']) for (u, v, d) in Actor_nw_undirected_weight.edges(data=True)]
edges =  [(u, v) for (u, v, d) in Actor_nw_undirected_weight.edges(data=True)]

# Initialise layout and plot nodes
pos = nx.circular_layout(Actor_nw_undirected_weight)
nx.draw_networkx_nodes(Actor_nw_undirected_weight, pos=pos)

# edges
nx.draw_networkx_edges(Actor_nw_undirected_weight, pos, edgelist=edges, edge_color=weights, width=5)

# edge labels
nx.draw_networkx_labels(Actor_nw_undirected_weight, pos, font_size=15, font_family='sans-serif')
edge_labels = nx.get_edge_attributes(Actor_nw_undirected_weight,'weight')
nx.draw_networkx_edge_labels(Actor_nw_undirected_weight, pos=pos, edge_labels=edge_labels)

fig = plt.gcf() # Get current figure (Get graph drawn by networkx)
fig.set_size_inches(10, 10)
plt.show()
````

### Density
Describe the portion of the **potential connections** (i.e. the person could know the others) in a network that are **actual connections** (i.e. the person knows). The density is 0 for a graph without edges and 1 for a complete graph. The denominator of equation (2a) and (2b) is the maximum number of edges in the graph $G$.


  - **Undirected**: since edge $e_{ij}$ is the same as $e_{ji}$, here $n$ is the number of nodes and $m$ is the number of edges in undirected graph $G$.
  $$d_{undirected} = \frac{Actual\_Connections}{Potential\_Connections} = \frac{m}{(\frac{n(n-1)}{2})} \tag{1a}$$

  - **Directed**: here $n$ is the number of nodes and $m$ is the number of edges in directed graph $G$.
  $$d_{directed} = \frac{Actual\_Connections}{Potential\_Connections} = \frac{m}{n(n-1)} \tag{1b}$$

````python
## Density for undirected graph
print("Density of the undirected graph:", nx.density(Actor_nw_undirected))
## Density for directed graph
print("Density of the directed graph:", nx.density(Actor_nw_directed))
````

### Clustering Coefficient

A measure of the degree to which nodes in a graph tend to cluster together. Here we will only discuss the local clustering coefficient, since we care which node in a network has a tendency to from clusters. **Local Clustering Coefficient** of a node in a network is the fraction of pairs of the node’s neighbours that are adjacent to each other.

 - **Undirected-unweighted network**: the clustering of a node $u$ is the fraction of possible triangles through that node that exist, where $T(u)$ is the number of edges among the neighbors of node $u$. And $neighbor(u)$ is the number of neighbor nodes for node $u$.
    - If $neighbor(u)>1$ 
  $$c_{u} = \frac{\text{Actual Connections (neighbor nodes)}}{\text{Potential Connections (neighbor nodes)}} = \frac{T(u)}{(\frac{neighbor(u)(neighbor(u)-1)}{2})} \tag{2a}$$
    - else
    $$c_u = 0 \tag{2b}$$ 
  
See more [here](https://networkx.github.io/documentation/stable/reference/algorithms/generated/networkx.algorithms.cluster.clustering.html#networkx.algorithms.cluster.clustering)

### Average Clustering

Average clustering coefficient for the network, where $n$ is the number of nodes in the network $G$ and $c_u$ is the clustering coefficient of the node $u$ in $G$:

$$C = \frac{1}{n}\sum_{u \in G} c_u \tag{3}$$ 

````python
# Clustering coefficient for undirected & unweighted graph
nx.draw_networkx(Actor_nw_undirected)
print(nx.clustering(Actor_nw_undirected))
print(nx.average_clustering(Actor_nw_undirected))
````

### Average Shortest Path Length

The average shortest path length is the sum of the shortest path (-or distance, geodesic path) $d(u,v)$ between a set of nodes $S$ (assuming the length is zero if $v$ is not reachable from $v$) normalized by $n(n-1)$ where $n$ is the number of nodes in network $G$. 

If ```weight``` is None, every edge has weight/distance/cost 1. If a string, use this edge attribute as the edge weight. Any edge attribute not present defaults to 1.

$$<d> =\sum_{u,v \in S} \frac{d(u, v)}{n(n-1)} \tag{5}$$ 

````python
nx.draw_networkx(Actor_nw_undirected)
avg_shortest_path_length = nx.average_shortest_path_length(Actor_nw_undirected)
print('Average shortest path of the unweighted & undirected actors graph: {}'.format(avg_shortest_path_length))
````

### Prominance - Centrality

**Centrality**: check which node (i.e. friend) is involved in many edges (i.e. relationship)

- **Degree Centrality**: the fraction of nodes it is connected to. Degree of node $u$ divided by the maximal possible degree (i.e. number of nodes in the graph subtract by 1)

  $$C_D(u) =\frac{deg(u)}{n-1}\tag{6}$$

- **Betweenness Centrality**: how many pairs of individuals would have to go through you in order to reach one another in the minimum number of hops. Betweenness centrality of a node $v$ is then defined as the sum of the fraction of all-pairs shortest paths that pass through $u$. Here $V$ is the set of nodes, $\sigma(s, t)$ is the number of shortest $(s,t)$-paths, and $\sigma(s, t|u)$ is the number of those paths passing through some node $u$ other than $s,t$.
$$C_B(u) =\sum_{s,t \in V} \frac{\sigma(s, t|u)}{\sigma(s, t)} \tag{7a}$$

     - **Normalized**: $n$ is the number of nodes in the graph

  $$C^{'}_B(u) = \frac{C_B(u)}{(\frac{(n-1)(n-2)}{2})} \tag{7b}$$

- **Closeness Centrality**: focus on how close a node $u$ is to all the other nodes in the network. It is based on the length of the average shortest path between a node and all the other nodes (i.e. $k-1$ reachable nodes) in the graph. Here $d(u, v)$ is the shortest-path distance between $u$ and $v$. 
  $$C_C(u) = \frac{1}{\sum_{v=1}^{k-1} d(u, v)} \tag{8a}$$

     - **Normalized**: $k-1$ is the number of nodes that can reach $u$. 

  $$C^{'}_C(u) = \frac{k - 1}{\sum_{v=1}^{k-1} d(u, v)} \tag{8b}$$

````python
nx.draw_networkx(Actor_nw_undirected)
print('Calculate the degree centrality per node:')
nx.degree_centrality(Actor_nw_undirected)
````

````python
print('Calculate the Betweenness Centrality per node:')
print(nx.betweenness_centrality(Actor_nw_undirected, normalized=False))
print('Calculate the Normalized Betweenness Centrality per node:')
print(nx.betweenness_centrality(Actor_nw_undirected, normalized=True))
````

````python
print('Calculate the Normalized Closeness Centrality per node:')
print(nx.closeness_centrality(Actor_nw_undirected)) # Networkx provides only a normalized implementation of this measure
````

### Prominance - Prestige

**Prestige**: Take the direction of arc into account (i.e. directed graph)
- **Input domain**: The input domain of a vertex in a directed network is the number or percentage of all other vertices that are connected by a path to this vertex.

- **Proximity Prestige**: Measure based on the distance of input domains. Input domain of a node in a directed graph is the number or percentage of all other nodes that are connected by a path to this node. Direct nominations (i.e. choices) should count more than indirect nominations. Here $n$ is the number of nodes in the graph and $d(i, u)$ is the distance between node $i$ and node $u$.
   $$C_P(u) =\frac{\text{fraction of all nodes that are in u's input domain} }{\text{average distance from u to other nodes in the input domain}} = \frac{\frac{|I_u|}{(n-1)}}{\frac{\sum_{i \in I_u} d(i, u)}{|I_u|}}$$

- **Page Rank**: more prestigious if a node has lots of other prestigious people in the input domain. [Check more here](https://networkx.github.io/documentation/stable/reference/algorithms/link_analysis.html)

````python
nx.draw_networkx(Actor_nw_directed)
print('Calculate the Input domains per node in a directed graph:')

def input_domain(G):
    node_2_input_domain = {node: len(nx.ancestors(G, node)) for node in G.nodes}
    return node_2_input_domain
                                     
print(input_domain(Actor_nw_directed))

def proximity_prestige(G):
    # Calculate fraction of all vertices that are in i's input domain
    node_2_inp_domain = input_domain(G)
    node_2_frac_input_domain = {node: inp_domain/len(G.nodes) for node, inp_domain in node_2_inp_domain.items()}
    
    # Calculate average distance from i to vertex in input domain
    node_2_avg_distance = {}
    for node in G.nodes:
        shortest_path_lengths = [nx.shortest_path_length(G, source=ancestor, target=node) for ancestor in nx.ancestors(G, node)]
        if len(shortest_path_lengths) > 0:
            node_2_avg_distance[node] = sum(shortest_path_lengths)/len(shortest_path_lengths)
        else:
            node_2_avg_distance[node] = 0
    
    #Calculate proximity prestige per node
    node_2_proximity_prestige = {}
    for node in node_2_frac_input_domain:
        if node_2_avg_distance[node] > 0:
            node_2_proximity_prestige[node] = node_2_frac_input_domain[node]/node_2_avg_distance[node]
        else:
            node_2_proximity_prestige[node] = 0
    
    return node_2_proximity_prestige

print('Calculate the proximity prestige per node in a directed graph:')
print(proximity_prestige(Actor_nw_directed))

print('Calculate the PageRank per node in a directed graph:')
print(nx.pagerank(Actor_nw_directed))
````

## Case-Study: Maintenance of the RNV network

Your objective is to identify key tram stops in the RNV network that require maintenance based on their network centrality measures.

Background: Tram networks are complex systems of inter-connected stops, where the maintenance of a single stop can have a ripple effect on the entire network. Some stops act as hubs connecting multiple lines, while others are peripheral stops serving a limited number of routes. Therefore, it is essential to identify the most important stops in the network, which are likely to have a significant impact on the overall performance of the system. Maintenance of these stops should be prioritized.

![rnv_graph.png](https://drive.google.com/uc?id=1Jj77a6-qIxaY06nTM_BA_qpPF1DmArH2)

The RNV network contains tram stops and their connections in the network. Each stop is represented by a node, and the connections between them are represented by edges.
The network was created based on the data from the [RNV open data portal](https://www.opendata-oepnv.de//ht/de/organisation/verkehrsunternehmen/rnv/openrnv/datensaetze?tx_vrrkit_view%5Bdataset_name%5D=soll-fahrplandaten-rnv&tx_vrrkit_view%5Baction%5D=details&tx_vrrkit_view%5Bcontroller%5D=View).

Build a network graph using the tram stop dataset, where each stop is a node, and each connection between stops is an edge.

````python
import networkx as nx
# Load the graph and exclude the weights.
G_rnv = nx.read_edgelist("/content/rnv_connected_stations.txt", create_using=nx.Graph(), delimiter='\t', data=False)
print('Number of stops (nodes): ', len(G_rnv.nodes))
print('Number of connections (edges): ', len(G_rnv.edges))

print("Density of the RNV graph:", nx.density(G_rnv))
````

## Plot the RNV graph

Plot the graph to get a first impression of the network.

````python
import matplotlib.pyplot as plt

nx.draw_networkx(G_rnv, font_size=30)
fig = plt.gcf()
fig.set_size_inches(50, 50)
plt.show()
````

## Maintenance of stops that are connected to many other stops.

A stop that has many connections with other stops in the network plays a crucial role in facilitating passenger transfers and ensuring efficient tram operations. If such a stop is not maintained properly, it can lead to disruptions in tram services, delays, and increased operating costs.

Your task is to identitify five stops that are connected to many other stops.

1) Which measure do you use to identify these stops?
2) Calculate the measure and find the 5 most connected stops.
3) Plot the graph and visualize the 5 most connected stops.

````python
# We have to calculate the degree centrality and filter for the 5 most important stops.
top_degree_scores = {id: round(btw_central,5) for id, btw_central in sorted(nx.degree_centrality(G_rnv).items(), key=lambda item: item[1], reverse=True)[:5]}
print('Top 10 based on the degree centrality score:\n', top_degree_scores)

# Get node dict and remove the label of all irrelevant nodes
node_dict = {v: v if v in top_degree_scores.keys() else '' for v in G_rnv.nodes()}

pos = nx.spring_layout(G_rnv, scale=5, seed=42)
# Emphasize relevant nodes in size and color
node_color_size = [top_degree_scores[v] * 10000 if v in top_degree_scores else 5 for v in G_rnv.nodes()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_rnv, pos=pos, with_labels=True,
                 node_color=node_color_size,
                 node_size=node_color_size, labels=node_dict )
plt.show()
````

## Maintenance of stops that are important for the reliability of the tram service.

A stop that lies on many of the shortest paths between other stops in the network plays a crucial role in facilitating passenger transfers and ensuring efficient tram operations. If such a stop is not maintained properly, it can lead to disruptions in tram services, delays, and increased operating costs. By prioritizing maintenance of these critical stops, the overall reliability and efficiency of the tram system can be improved, resulting in a better user experience and reduced operating costs.

Your task is to identitify five stops that are important for the reliability of the tram service.

1) Which measure do you use to identify these stops?
2) Calculate the measure and find the 5 most important stops.
3) Plot the graph and visualize the 5 most important stops.

````python
# We have to calculate the Betweeness centrality and filter for the 5 most important stops.
top_betweeness_scores = {id: round(btw_central,5) for id, btw_central in sorted(nx.betweenness_centrality(G_rnv, normalized =True).items(), key=lambda item: item[1], reverse=True)[:5]}
print('Top 5 based on the betweenness centrality score:\n', top_betweeness_scores)

# Get node dict and remove the label of all irrelevant nodes
node_dict = {v: v if v in top_betweeness_scores.keys() else '' for v in G_rnv.nodes()}

pos = nx.spring_layout(G_rnv, scale=5, seed=42)
# Emphasize relevant nodes in size and color
node_color_size = [top_betweeness_scores[v] * 1000 if v in top_betweeness_scores else 5 for v in G_rnv.nodes()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_rnv, pos=pos, with_labels=True,
                 node_color=node_color_size,
                 node_size=node_color_size, labels=node_dict )
plt.show()
````
